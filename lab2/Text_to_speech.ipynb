# Text to speech

Технология текст в речь (Text-to-Speech, TTS) представляет собой процесс преобразования письменного текста в звучащую речь. Эта технология широко используется в различных приложениях и устройствах для предоставления аудиальной информации пользователям. Системы TTS используют компьютерные алгоритмы и синтетические голоса для генерации речи на основе входного текста.

Технология TTS имеет широкий спектр применений, включая функции доступности для людей с ограниченными возможностями зрения, навигационные системы, голосовых ассистентов, таких как Siri и Alexa, приложения для обслуживания клиентов и многое другое. Она позволяет компьютерам и устройствам общаться с пользователями на уровне устной речи, что делает информацию и услуги более доступными и удобными для пользователей.

```python
!pip3 install -U scipy

!git clone https://github.com/jnordberg/tortoise-tts.git
%cd tortoise-tts
!pip3 install -r requirements.txt
!pip3 install transformers==4.19.0 einops==0.5.0 rotary_embedding_torch==0.1.5 unidecode==1.3.5
!python3 setup.py install

import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F

import IPython

from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_audio, load_voice, load_voices

tts = TextToSpeech()

import os
from google.colab import files
```

# Cоздание голоса

```python
# Озвучиваемый текст
text = "Science is important to world peace in many ways. On one hand, scientists have helped to develop many of the modern tools of war. On the other hand, they have also helped to keep the peace through research, which has improved life for people."

# пресеты качества, можно экспериментировать. Варианты: {"ultra_fast", "fast" (default), "standard", "high_quality"}.
preset = "high_quality"

# Название нового голоса
CUSTOM_VOICE_NAME = "Bella"

# Загрузка семплов голоса. WAV формат
custom_voice_folder = f"tortoise/voices/{CUSTOM_VOICE_NAME}"
os.makedirs(custom_voice_folder)
for i, file_data in enumerate(files.upload().values()):
  with open(os.path.join(custom_voice_folder, f'{i}.wav'), 'wb') as f:
    f.write(file_data)
```

# генерация аудио

```python
# Генерация аудио, не забудьте указать имя созданного вами голоса
voice_samples, conditioning_latents = load_voice("angie")
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                          preset=preset)
torchaudio.save(f'generated-{CUSTOM_VOICE_NAME}.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio(f'generated-{CUSTOM_VOICE_NAME}.wav')
```

При выборе одного голоса с сайта elevenlabs и загрузив его в модель tortoise было проведено сравнение работы различных моделей. С помощью сети tortoise получается более “человечный” голос, чем при применении модели с сайта elevenlabs. Предположительно, более “человечный” голос связан с различиями в данных, на которых обучались эти модели, более подходящей архитектурой модели, более эффективными методами обучения и более точной настройкой гиперпараметров.
